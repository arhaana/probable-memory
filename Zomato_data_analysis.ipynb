```python
# prompt: suggests that a majority of the restaurants do not accept online orders.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataframe = pd.read_csv("/content/Zomato data .csv")
print(dataframe.head())
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())




dataframe.info()

sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red",size=20)

max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print("Restaurant(s) with the maximum votes:")
print(restaurant_with_max_votes)

# Count the number of restaurants that accept online orders
online_order_counts = dataframe['online_order'].value_counts()

# Print the counts
print(online_order_counts)

# Calculate the percentage of restaurants that do not accept online orders
no_online_percentage = (online_order_counts['No'] / online_order_counts.sum()) * 100

print(f"Percentage of restaurants that do not accept online orders: {no_online_percentage:.2f}%")
# This code snippet calculates and prints the percentage of restaurants that do not accept online orders.
# It suggests that a majority of the restaurants might not accept online orders if the percentage is significantly high.
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Ratings Distribution")
plt.show()
```
